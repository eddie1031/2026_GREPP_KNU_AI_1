```python
from langchain_core.runnables import RunnableLambda

append_ten = lambda x: x + 10
# multiply_two = lambda x: x * 2

run_append = RunnableLambda(append_ten) # Runnable
run_multiply = RunnableLambda(lambda x: x * 2) # Runnable

append_result = run_append.invoke(5)

print(f'append_result = {append_result}')

multiply_result = run_multiply.invoke(append_result)

print(f'multiply_result = {multiply_result}')


other_result = run_multiply.invoke(
    run_append.invoke(10)
)


print(f'other_result = {other_result}')
```

```python
import asyncio
import time

import nest_asyncio

nest_asyncio.apply()

from langchain_core.runnables import RunnableLambda


async def append_ten(x: int) -> int:
    print(f'{x} 를 받아서 append_ten 시작!')
    await asyncio.sleep(2)
    print(f'{x} 를 받아서 append_ten 종료!')
    return x + 10

run_async_append = RunnableLambda(append_ten)

# invoke   /  ainvoke
# batch    /  abatch
# stream   /  astream

async def main():

    start = time.time()

    results = await asyncio.gather(
        run_async_append.ainvoke(5),
        run_async_append.ainvoke(3),
        run_async_append.ainvoke(8),
    )

    end = time.time()

    print('결과 ' , results)

    print('총 실행시간', round(end - start, 2), "초")


asyncio.run(main())
```

```python
from langchain_core.runnables import RunnableLambda

import time

def print_str(data: str):
    for _ in data:
        time.sleep(0.3)
        yield _

run_print_str = RunnableLambda(print_str)

for _ in run_print_str.stream("안녕하세요 여러분! 반갑습니다! stream() 메서드를 살펴봅시다."):
    print(_, end="", flush=True)


```
