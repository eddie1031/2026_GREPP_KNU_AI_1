```python
from langchain_groq import ChatGroq

from dotenv import load_dotenv
import os

load_dotenv()

API_KEY = os.getenv("GROQ_API_KEY")

llm = ChatGroq(
    model="openai/gpt-oss-20b",
    api_key=API_KEY
)

result = llm.invoke("인공지능이 뭐야?")

result.pretty_print()

```
