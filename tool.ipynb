```python
def add(a: int, b:int ) -> int:
    """두 수를 더한다.
    
    Args:
        a (int): 첫 번째 숫자
        b (int): 두 번째 숫자
    
    Returns:
        int: 두 수의 합
    """
    return a + b


```

```python
from langchain_core.tools import tool

@tool
def get_weather(city: str) -> str:
    """도시의 현재 날씨를 반환합니다.

    Args:
        city: 도시의 이름
    
    Returns:
        str: 도시의 현재 날씨 정보
    """
    mock_data = {
        '서울': '맑음, 22도',
        '대전': '흐림, 23도',
        '천안': '정말 맑음, 21도',
    }

    return mock_data.get(city, '날씨 정보 없음')


result = get_weather.invoke({'city': '천안'})

print(result)

print("도구 이름 : ", get_weather.name)
print("도구 설명 : ", get_weather.description)

```

```python
from langchain_groq import ChatGroq
from dotenv import load_dotenv

load_dotenv()

model = ChatGroq(
    model = "openai/gpt-oss-20b",
    temperature=0,
)

tools = [ get_weather ]

model_with_tools = model.bind_tools(tools)

model_with_tools.invoke("대전의 날씨를 알려줘")


```
